```python
import cv2
```

```python
# import numpy as np
# import sys
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("difficulty_expert_p1.png")
```

```python
# making both height and width of the same dimension which is 9 * 28
image = cv2.resize(image, (252,252), interpolation = cv2.INTER_LINEAR)
```

```python
h,w,d = image.shape
box = 28
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 0.6
color = (255, 0, 0)
thickness = 1
```

```python
for i in range(0,h,box):
    for j in range(0,w,box):
        org = (i+7, j+21)
        text = '5'
        image = cv2.putText(image, text, org, font, fontScale, color, thickness, cv2.LINE_AA)
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5)
plt.imshow(image)
# plt.show()
plt.savefig('foo.png')
```
